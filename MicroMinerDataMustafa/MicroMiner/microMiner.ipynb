```python
from google.colab import drive
import os
import re
import numpy as np

drive.mount('/content/drive')
os.chdir('/content/drive/My Drive/MicroMinerDataMustafa/MicroMiner/newDir')
def process_files_with_model(model_name, get_embedding_func, stop_words, aSIF, allWords, typedClasses):
    embadDict = {}
    wListClass = {}
    rx = re.compile(r'(?<=[a-z])(?=[A-Z])')

    for elem in typedClasses:
        filelines = readFile(elem.lower() + '.java')
        strings = []
        for s in filelines:
            sentence = " ".join(re.findall("[a-zA-Z]+", s))
            words = sentence.split()
            for w in words:
                if w not in stop_words:
                    strings.append(w)

        nstrings = [re.findall('.[^A-Z]*', string) for string in strings]
        all_words = []
        for ns in nstrings:
            all_words = all_words + ns

        treatedW = []
        embads = []
        w = []
        for a in all_words:
            treatedW.append(getLemm(a.lower()))

        for t in treatedW:
            if t in v2w_model and t not in stop_words and len(t) > 1:
                w.append(t)
                coefSIF = aSIF / (aSIF + allWords[t])
                embads.append(get_embedding_func(t))

        wListClass[elem] = w
        embads = np.array(embads)

        if len(embads) > 1:
            meanmatrix = embads.mean(0)
            embadDict[elem] = meanmatrix
        else:
            embadDict[elem] = embads

    return embadDict
```
